Environment Setup & Data Loading

```python
# Task 1: Import NumPy, Pandas, Matplotlib.pyplot and Seaborn.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Task 2: Load train.csv into a Pandas DataFrame using pd.read_csv().
# Load the dataset
df = pd.read_csv("Titanic-Dataset.csv")

# Display the first 5 rows
print(df.head())
```

Data Inspection

```python
# Task 3: Display first 10 rows using df.head(10).
# Display first 10 rows
print(df.head(10))
```

```python
# Task 4: Display last 5 rows using df.tail(5).
# Display last 5 rows
print(df.tail(5))
```

```python
# Task 5: Check dataset dimensions using df.shape (Expected: 891 x 12).
# Check the number of rows and columns
print(df.shape)
```

```python
# Task 6: Display dataset information using df.info().
# Display dataset information
df.info()
```

```python
# Task 7: Generate summary statistics using df.describe(); record mean, minimum and maximum values of Age and Fare.
# Summary statistics
print(df.describe())
```

```python
# Task 8: Identify missing values using df.isnull().sum().
# Count missing values in each column
print(df.isnull().sum())
```

Data Manipulation & Cleaning

```python
# Task 9: Fill missing Age values with the median.
# Fill missing Age values with the median
df["Age"].fillna(df["Age"].median(), inplace=True)

# Verify missing values
print(df["Age"].isnull().sum())
```

```python
# Task 10: Filter surviving passengers (Survived == 1).
# Filter passengers who survived
survivors = df[df["Survived"] == 1]

# Display first 5 surviving passengers
print(survivors.head())

# Display total number of survivors
print("Total Survivors:", len(survivors))
```

```python
# Task 11: Convert Fare column to a NumPy array.
# Convert Fare column to NumPy array
fare_array = df["Fare"].to_numpy()

# Display the array
print(fare_array)
```

```python
# Task 12: Calculate maximum and average Fare using np.max() and np.mean().
# Maximum Fare
print("Maximum Fare:", np.max(fare_array))

# Average Fare
print("Average Fare:", np.mean(fare_array))
```

Data Visualization

```python
# Task 13: Plot histogram of Age.
# Plot Histogram of Age
plt.figure(figsize=(8,5))
plt.hist(df["Age"], bins=20, edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
# Task 14: Plot bar chart of Survival Count.
# Plot Bar Chart of Survival Count
survival_count = df["Survived"].value_counts()

plt.figure(figsize=(6,4))
plt.bar(survival_count.index, survival_count.values)
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")
plt.xticks([0, 1])
plt.show()
```

```python
# Task 15: Plot scatter graph of Age vs Fare.
# Plot Scatter Graph of Age vs Fare
plt.figure(figsize=(8,5))
plt.scatter(df["Age"], df["Fare"])
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
